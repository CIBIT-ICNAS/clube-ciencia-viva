## Etapa 2: Questão

Ainda se lembram do vossa questão científica? Segundo os nossos apontamentos da última sessão a vossa questão de investigação é a seguinte: 

> Qual será o nível de condutância da pele numa situação tensa e numa relaxada?


## Etapa 3: Hipótese

No entanto, é necessário formular as vossas hipóteses. A hipótese nula ($H_0$) que afirma, geralmente, que não existe relação entre dois fenómenos medidos e a hipótese alternativa ($H_1$)  que testa o contrário, ou seja, que existe relação.

As hipóteses que vocês desenvolveam-se encontram-se abaixo:


> $H_1$: O nível de condutância da pele é alterado pelo tipo de vídeo que estamos a ver (não há diferenças entre vídeos tensos e relaxados no SCL) 


> $H_0$: O nível de condutância da pele não é alterada pelo tipo de vídeo que estamos a ver (não há diferenças entre vídeos tensos e relaxados no SCL)



## Etapa 4: Metodologia

Por fim, vamos à metodologia que menciona o que se irá fazer, o que é necessário medir, quantas pessoas vão ser recrutadas e quais as usas características. Abaixo encontram-se o que foi definido anteriormente.

### Técnica

> Condutância da pele



### População 

> Indivíduos que frequentam a escola Martim de Freitas, com idades compreendidas entre os 10-18 anos.



### Critérios de inclusão/exclusão
> **Exclusão:** Alcoólicos, Fumadores, Consumidores de drogas que afetem o sistema nervoso, pacientes com doenças dermatológicas.

> **Tamanho da amostra:**  5/10

### Protocolo experimental
**Tarefa:** visualização de vídeos (definir duração em minutos)

> **Variáveis dependentes:** Condutância da pele (nível de condutância da pele)

> **Variáveis independentes:** Factor “Tipo de vídeo” (2 condições experimentais: calmo vs tenso)

> **Variáveis de controlo:** Questionário Ansiedade Estado (STAI); Respostas dos participantes relativamente a tipo de vídeo (activação/intensidade do estímulo; valência)

#### Procedimentos de aplicação

> 1.   lavar mãos, sentar participantes, 
> 2.   aplicação eléctrodos: mesmo local entre participantes - definir local (2 eminências da palma da mão)
> 3.   verificar distância entre eléctrodos
> 4.   prender com fita adesiva para eléctrodo não descolar com suor
> 5.   minutos iniciais para SCL atingir baseline; 
> 6.   distância participante - ecrã…



#### Gestão de dados: 
> 1.   anonimização (separação nome-código); 
> 2.   registo de dados em que suporte (open source preferencialmente)

#### Outputs de Investigação
> Dados; código de análise; posters; mini-artigo

#### Plano de Análise de dados:
> 1.   Contrastes testes-t
> 2.   correlação variáveis SCL com arousal/valência (ou usar esses factores como co-variáveis)
